```python


def jugar_torneo():
    opciones = ["piedra", "papel", "tijera"]
    puntos_usuario = 0
    puntos_pc = 0
    
    print("--- üèÜ INICIO DEL TORNEO (Al mejor de 3) üèÜ ---")

    while puntos_usuario < 2 and puntos_pc < 2:
        # 1. Elecci√≥n de la PC
        pc = random.choice(opciones)
        
        # 2. Tu elecci√≥n (con validaci√≥n)
        usuario = random.choice(opciones)
        
        if usuario not in opciones:
            print("‚ùå Opci√≥n no v√°lida. Intenta de nuevo.")
            # TODO: ¬øQu√© palabra clave usamos para saltar al inicio del bucle sin avanzar?
            continue 
            
        print(f"üÜö T√∫: {usuario} | PC: {pc}")
        
        # 3. L√≥gica para determinar el ganador de la ronda
        if usuario == pc:
            print("üòê ¬°Empate! Nadie suma puntos.")
        elif (
            # TODO: Escribe aqu√≠ la condici√≥n l√≥gica COMPLETA para que gane el usuario
            # Pista: (usuario es piedra Y pc es tijera) O (...) O (...)
            (usuario == "piedra" and pc == "tijera") or
            (usuario == "papel" and pc == "piedra") or
            (usuario == "tijera" and pc == "papel")
        ):
            print("‚ú® ¬°Ganaste la ronda!")
            puntos_usuario += 1
        else:
            print("ü§ñ La PC gana la ronda.")
            puntos_pc += 1
            
        print(f"   Marcador -> T√∫: {puntos_usuario} - PC: {puntos_pc}\n")

    # 4. Resultado final
    # TODO: Escribe un if/else para felicitar al ganador real
    if puntos_usuario > puntos_pc:
        print("üéâ ¬°FELICIDADES! Ganaste el torneo.")
    else:
        print("üíÄ Fin del juego. La PC gan√≥ el torneo.")

# Ejecutar el juego
# jugar_torneo()
```
